```python
query = """
SELECT 
    Country,
    SUM(Revenue) as revenue
FROM sales
GROUP BY Country
ORDER BY revenue DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT 
    CustomerID,
    SUM(Revenue) as total_spent
FROM sales
GROUP BY CustomerID
ORDER BY total_spent DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT 
    Description,
    SUM(Revenue) as revenue
FROM sales
GROUP BY Description
ORDER BY revenue DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT 
    strftime('%Y-%m', InvoiceDate) as month,
    SUM(Revenue) as revenue
FROM sales
GROUP BY month
ORDER BY month
"""

pd.read_sql(query, conn)
```
